### Groupby

```python
def range_fun(x):
  return x.max()-x.min()

df.groupby('company').agg(range_fun)
```

```python
# inner join

import pandas as pd

# Create the first dataframe
df1 = pd.DataFrame({
    'id': range(1, 11),
    'value1': [f'A{i}' for i in range(1, 11)],
    'common_key': [i for i in range(3, 13)]
})

# Create the second dataframe
df2 = pd.DataFrame({
    'id': range(5, 15), # Overlap in 'id' with df1
    'value2': [f'B{i}' for i in range(5, 15)],
    'common_key': [i for i in range(5, 15)] # Overlap in 'common_key' with df1
})

# Perform inner join on the 'common_key' column
inner_join_df = pd.merge(df1, df2, on='common_key', how='inner')

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nInner Join Result:")
print(inner_join_df)
```

```python
# Perform inner join on the 'common_key' column
outer_join_df = pd.merge(df1, df2, on='common_key', how='outer')
outer_join_df
```

```python
left_join_df = pd.merge(df1, df2, on='common_key', how='left')
left_join_df
```

```python
right_join_df = pd.merge(df1, df2, on='common_key', how='right')
right_join_df
```

```python
df_concat = pd.concat([df1, df2], axis=0) # add row
df_concat
```

```python
df_concat = pd.concat([df1, df2], axis=1) # add col
df_concat
```

### Handling Categorical variable

Nominal = can not be order. eg: gender, fruits

ordinal = can be order(can be arragne in assending or descending order). eg: customer satisfaction level, econmic status

```python
# Sample data
data = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red', 'Yellow']})

# One-hot encoding
encoded_data = pd.get_dummies(data, columns=['Color']).astype(int)

# One-hot encoding with first category dropped
encoded_data_drop_first = pd.get_dummies(data, columns=['Color'], drop_first=True).astype(int)

print("One-hot encoding:")
print(encoded_data)

print("\nOne-hot encoding with first category dropped:")
print(encoded_data_drop_first)
```

```python
from sklearn.preprocessing import LabelEncoder
# Label encoding
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(data['Color'])

# Adding encoded labels to the DataFrame
data['Encoded_Color'] = encoded_labels

print(data)
```

```python
# Sample data
fruits = pd.DataFrame({'fruits': ['Apple', 'Banana', 'Orange', 'Apple', 'Mango', 'Banana']})

# Define mapping dictionary for label encoding
fruit_mapping = {
    'Apple': 0,
    'Banana': 1,
    'Orange': 2,
    'Mango': 3
}

# Apply label encoding
fruits['fruits_label'] = fruits['fruits'].map(fruit_mapping)

print(fruits)
```

```python
#target encoding
# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Target': [100, 200, 150, 300, 250, 350, 125, 225]
})

# Compute the mean target value for each category
target_means = data.groupby('Category')['Target'].mean()

# Apply target encoding
data['Category_encoded'] = data['Category'].map(target_means)

print(data)
```

True if all of a set of comparision are True

True if any one of a set of comparision is TRUE

AND:- AND operator allows the existence of multiple conditions is an SQL statement WHERE clause. AND makes it necessary for all conditions to be true

OR:- combine multiple conditions in a SQL statement

BETWEEN:- to search for values that are within a set of value that are within a set of values, given the minimum value and the maximum value.
